### Visualizing Linearity using Scatter Plots

```python
# Visualise the linearity of independent and dependent variables using scatter plots
'''Create subplots in 2 row and 3 columns'''
def show_linearity_subplots_for(df):
    fig, ax = plt.subplots(2,3)
    fig.set_figheight(10)
    fig.set_figwidth(20)
    fig.subplots_adjust(left=.2, bottom=None, right=None, top=None, wspace=.5, hspace=.5)
    plt1 = plt.subplot(2,3,1)
    plt2 = plt.subplot(2,3,2)
    plt3 = plt.subplot(2,3,3)
    plt4 = plt.subplot(2,3,4)
    plt5 = plt.subplot(2,3,5)
    plt6 = plt.subplot(2,3,6)
    

    plt1.scatter(df[df.columns[0]], df[df.columns[6]])
    plt1.set_xlabel(df.columns[0])
    plt1.set_ylabel(df.columns[6])
    plt1.grid(axis='both', alpha=.25)

    plt2.scatter(df[df.columns[1]], df[df.columns[6]])
    plt2.set_xlabel(df.columns[1])
    plt2.set_ylabel(df.columns[6])
    plt2.grid(axis='both', alpha=.25)

    plt3.scatter(df[df.columns[2]], df[df.columns[6]])
    plt3.set_xlabel(df.columns[2])
    plt3.set_ylabel(df.columns[6])
    plt3.grid(axis='both', alpha=.25)

    plt4.scatter(df[df.columns[3]], df[df.columns[6]])
    plt4.set_xlabel(df.columns[3])
    plt4.set_ylabel(df.columns[6])
    plt4.grid(axis='both', alpha=.25)

    plt5.scatter(df[df.columns[4]], df[df.columns[6]])
    plt5.set_xlabel(df.columns[4])
    plt5.set_ylabel(df.columns[6])
    plt5.grid(axis='both', alpha=.25)

    plt6.scatter(df[df.columns[5]], df[df.columns[6]])
    plt6.set_xlabel(df.columns[5])
    plt6.set_ylabel(df.columns[6])
    plt6.grid(axis='both', alpha=.25)
```

```python
show_linearity_subplots_for(df)
```

## Data Normalization

```python
df_norm = (df-np.mean(df, axis = 0))/np.std(df, axis = 0)
```

```python
df_norm.head()
```

```python
show_linearity_subplots_for(df_norm)
```
